```python
from src.data.movielens1M import LoadMovielens1M, PreprocessMovielens1M
from src.data.split import random_split, per_user_split
from src.loss.error_log import get_log_per_test_size
```

```python
load_rawML1M = LoadMovielens1M(path='datasets/ml-1m')
raw_datas = load_rawML1M.load_data()

preprocessML1M = PreprocessMovielens1M(raw_datas)
preprocess_datas = preprocessML1M.apply_prepocess()

ratings, users, movies, histories = preprocess_datas
```

```python
TEST_SIZE = [0.1, 0.15, 0.2, 0.25]
```

```python
random_log = get_log_per_test_size(random_split, ratings, TEST_SIZE)
```

```python
per_user_log = get_log_per_test_size(per_user_split, ratings, TEST_SIZE)
```

```python
import matplotlib.pyplot as plt

plt.plot(TEST_SIZE, random_log[:, 0], label="random")
plt.plot(TEST_SIZE, per_user_log[:, 0], label="per_usr")
plt.title('RMSE per test_size with val data')
plt.legend()
plt.show()
plt.close()
```

```python
plt.plot(TEST_SIZE, random_log[:, 1], label="random")
plt.plot(TEST_SIZE, per_user_log[:, 1], label="per_usr")
plt.title('RMSE per test_size with test data')
plt.legend()
plt.show()
plt.close()
```
